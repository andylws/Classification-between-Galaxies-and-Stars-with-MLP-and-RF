```python
import os.path
import numpy as np
from PIL import Image
from glob import glob
```

```python
import csv

def csvWriter(fil_name, numlist):
    with open(fil_name+'.csv', 'w', newline='') as csvfile:
        writer = csv.writer(csvfile, delimiter=',')
        writer.writerows(numlist)
```

---

```python
# 28x28 pixels training dataset

galaxy_train_imglist_28 = []
star_train_imglist_28 = []

os.chdir('C:/Users/andyl/Desktop/Assignment10/archive/data/train/galaxy_28')
for file in glob('*.png'):
    img = np.array(Image.open(file))
    galaxy_train_imglist_28.append(img)
    
os.chdir('C:/Users/andyl/Desktop/Assignment10/archive/data/train/star_28')
for file in glob('*.png'):
    img = np.array(Image.open(file))
    star_train_imglist_28.append(img)
    
full_list_28 = []

for num in range(800):
    for num1 in range(5):
        full_list_28.append([0])
        for i in range(28):
            for j in range(28):
                full_list_28[num * 10 + num1].append(galaxy_train_imglist_28[num * 5 + num1][i][j][0])
                full_list_28[num * 10 + num1].append(galaxy_train_imglist_28[num * 5 + num1][i][j][1])
                full_list_28[num * 10 + num1].append(galaxy_train_imglist_28[num * 5 + num1][i][j][2])
    for num2 in range(5):
        full_list_28.append([1])
        for i in range(28):
            for j in range(28):
                full_list_28[num * 10 + num2 + 5].append(star_train_imglist_28[num * 5 + num2][i][j][0])
                full_list_28[num * 10 + num2 + 5].append(star_train_imglist_28[num * 5 + num2][i][j][1])
                full_list_28[num * 10 + num2 + 5].append(star_train_imglist_28[num * 5 + num2][i][j][2])
                
os.chdir('C:/Users/andyl/Desktop/Assignment10/archive/data/train')
csvWriter("train_rgb_28", full_list_28)
```

```python
# 28x28 pixels test dataset

galaxy_test_imglist_28 = []
star_test_imglist_28 = []

os.chdir('C:/Users/andyl/Desktop/Assignment10/archive/data/validation/galaxy_28')
for file in glob('*.png'):
    img = np.array(Image.open(file))
    galaxy_test_imglist_28.append(img)
    
os.chdir('C:/Users/andyl/Desktop/Assignment10/archive/data/validation/star_28')
for file in glob('*.png'):
    img = np.array(Image.open(file))
    star_test_imglist_28.append(img)
    
full_list_test_28 = []

for num in range(200):
    for num1 in range(5):
        full_list_test_28.append([0])
        for i in range(28):
            for j in range(28):
                full_list_test_28[num * 10 + num1].append(galaxy_test_imglist_28[num * 5 + num1][i][j][0])
                full_list_test_28[num * 10 + num1].append(galaxy_test_imglist_28[num * 5 + num1][i][j][1])
                full_list_test_28[num * 10 + num1].append(galaxy_test_imglist_28[num * 5 + num1][i][j][2])
    for num2 in range(5):
        full_list_test_28.append([1])
        for i in range(28):
            for j in range(28):
                full_list_test_28[num * 10 + num2 + 5].append(star_test_imglist_28[num * 5 + num2][i][j][0])
                full_list_test_28[num * 10 + num2 + 5].append(star_test_imglist_28[num * 5 + num2][i][j][1])
                full_list_test_28[num * 10 + num2 + 5].append(star_test_imglist_28[num * 5 + num2][i][j][2])
                
os.chdir('C:/Users/andyl/Desktop/Assignment10/archive/data/validation')
csvWriter("test_rgb_28", full_list_test_28)
```
