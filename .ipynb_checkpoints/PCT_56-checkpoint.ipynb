```python
import matplotlib.pyplot

class Perceptron:
    def __init__(self, input_dim, hidden_dim, output_dim, lr =0.1):
        self.w1 = np.random.normal(0.0, pow(input_dim, -0.5), (input_dim, hidden_dim))
        self.w2 = np.random.normal(0.0, pow(hidden_dim, -0.5), (hidden_dim, output_dim))
        self.h = np.zeros((1, hidden_dim))
        self.lr = lr
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim
        self.theta = 0
        
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def feedforward(self, x):
        a = x.astype(float)
        b = self.w1.astype(float)
        self.h = self.sigmoid(np.dot(a, b) - self.theta)
        return self.sigmoid(np.dot(self.h, self.w2) - self.theta)
    
    def feedforward_upto_hidden(self, x):
        return self.sigmoid(np.dot(x, self.w1) - self.theta)
    
    def bprop_w2(self, g, y):
        q = (-2) * (g - y) * y * (1 - y)
        return np.dot(self.h.reshape(self.hidden_dim, 1), q.reshape(1, self.output_dim))
    
    def bprop_w1(self, g, y, x):
        q1 = (-2) * (g - y) * y * (1 - y)
        q2 = np.dot(self.w2, q1)
        return np.dot(x.reshape(self.input_dim, 1), q2 * self.h * (1 - self.h).reshape(1, self.hidden_dim))
    
    def training(self, input, target):
        x = np.array(input).T
        y = self.feedforward(x)
        g = np.array(target).T
        
        self.w1 = self.w1 - self.lr * self.bprop_w1(g, y, x)
        self.w2 = self.w2 - self.lr * self.bprop_w2(g, y)
        
        
input_dim = 784
hidden_dim_100 = 100
hidden_dim_200 = 200
output_dim = 10
epoch_10 = 10
epoch_20 = 20

pct_100_10 = Perceptron(input_dim, hidden_dim_100, output_dim)
pct_100_20 = Perceptron(input_dim, hidden_dim_100, output_dim)
pct_200_10 = Perceptron(input_dim, hidden_dim_200, output_dim)
pct_200_20 = Perceptron(input_dim, hidden_dim_200, output_dim)

training_dataset_file = open("train_28.csv", 'r')

training_dataset_list = training_dataset_file.readlines()
training_dataset_file.close()
input_list = list()

# training with 100 hidden dimension for 10 epoch
for k in range(epoch_10):
    pct_100_10.lr = pct_100_10.lr * 0.8
    for i in training_dataset_list:
        all_values = i.split(',')
        inputs = (np.asfarray(all_values[1:]) / 255.0 * 0.99) + 0.01
        input_list.append(inputs)
        
        targets = np.zeros(output_dim) + 0.01
        targets[int(all_values[0])] = 0.99
        
        pct_100_10.training(inputs, targets)
        
# training with 100 hidden dimension
for k in range(epoch_20):
    pct_100_20.lr = pct_100_20.lr * 0.8
    for i in training_dataset_list:
        all_values = i.split(',')
        inputs = (np.asfarray(all_values[1:]) / 255.0 * 0.99) + 0.01
        input_list.append(inputs)
        
        targets = np.zeros(output_dim) + 0.01
        targets[int(all_values[0])] = 0.99
        
        pct_100_20.training(inputs, targets)

# training with 200 hidden dimension
for k in range(epoch_10):
    pct_200_10.lr = pct_200_10.lr * 0.8
    for i in training_dataset_list:
        all_values = i.split(',')
        inputs = (np.asfarray(all_values[1:]) / 255.0 * 0.99) + 0.01
        input_list.append(inputs)
        
        targets = np.zeros(output_dim) + 0.01
        targets[int(all_values[0])] = 0.99
        
        pct_200_10.training(inputs, targets)
        
# training with 200 hidden dimension
for k in range(epoch_20):
    pct_200_20.lr = pct_200_20.lr * 0.8
    for i in training_dataset_list:
        all_values = i.split(',')
        inputs = (np.asfarray(all_values[1:]) / 255.0 * 0.99) + 0.01
        input_list.append(inputs)
        
        targets = np.zeros(output_dim) + 0.01
        targets[int(all_values[0])] = 0.99
        
        pct_200_20.training(inputs, targets)
        
pct = [pct_100_10, pct_100_20, pct_200_10, pct_200_20]
```

```python
test_file = open("test_28.csv", 'r')

test_list = test_file.readlines()
test_file.close()
output_list = list()

hidden_dim = [100, 100, 200, 200]
epoch = [10, 20, 10, 20]

success = [0, 0, 0, 0]
failure = [0, 0, 0, 0]

print("28x28 pixels")

for i in range(4):
    for j in test_list:
        all_values = j.split(',')
        target = int(all_values[0])

        all_values_normalized = (np.asfarray(all_values[1:]) / 255.0 * 0.99) + 0.01
        prediction_list = pct[i].feedforward(np.asfarray(all_values_normalized))
        prediction = np.argmax(prediction_list)

        if target == prediction:
            success[i] = success[i] + 1
        else:
            failure[i] = failure[i] + 1
            
    print("Recognition error rate of perceptron with " + str(hidden_dim[i]) + " hidden dimension for " + str(epoch[i]) + " epoch: " + str(failure[i] / (success[i] + failure[i])))
```
